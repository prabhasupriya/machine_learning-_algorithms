```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
#data preprocessing
sample = ['Gender', 'Married','Education',
       'Self_Employed','Property_Area' ]
```

```python
for i in sample:
  data[i]= le.fit_transform(data[i])
```

```python
data.head()
```

```python
data.info()
```

```python
data['Dependents'] = data['Dependents'].fillna(0)
```

```python
data['Credit_History'] = data['Credit_History'].fillna(data['Credit_History'].mode())
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mean())
```

```python
data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mean())
```

```python
numcol = data.select_dtypes(include=['int64','float64']).columns
```

```python
for i in numcol:
  data[i]= data[i].fillna(data[i].mode()[0])
```

```python
data.isnull().sum()
```

```python
data['LoanAmount']= data.LoanAmount.fillna(data.LoanAmount.mean())
```

```python
data.info()
```

```python
data= data.drop(['Loan_ID'] , axis=1)
```

```python
x= data.iloc[0:,:-1].values
y = data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test=  train_test_split(x,y,test_size=0.2,random_state= 2)
```

```python
x_train
y_train
```

```python
#model training
from sklearn.linear_model import LogisticRegression #classification
```

```python
lgmodel = LogisticRegression()
```

```python
#train the dataset
lgmodel.fit(x_train,y_train)
```

```python
y_pred = lgmodel.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix,precision_score,recall_score
```

```python
print(f"the acuracy score {accuracy_score(y_test,y_pred)*100}")
```

```python
print(f"the confusion matrix {confusion_matrix(y_pred,y_test)}")
print(f"the confusion matrix {precision_score(y_pred,y_test,pos_label='Y')}")
print(f"the confusion matrix {recall_score(y_pred,y_test,pos_label='Y')}")
```
