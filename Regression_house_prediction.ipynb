```python
#model training
from sklearn.neighbors import KNeighborsRegressor # same for KNeighborsClassifier
```

```python
#data preprocessing step 3
from sklearn.preprocessing import LabelEncoder # label encoder converts the strings into the numbers
```

```python
le = LabelEncoder()
```

```python
sampledata =['mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
       'parking', 'prefarea', 'furnishingstatus']
```

```python
for i in sampledata:
  data[i]= le.fit_transform(data[i])
```

```python
data
```

```python
x= data.iloc[: ,1:].values #iloc menas integer location, integer location[row,colums] iloc[: , :] all rows and all columns
y = data['price'].values
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.2 , random_state =2)
```

```python
knnmodel = KNeighborsRegressor(n_neighbors=111) #if not ension the n_neighbors it take default value as 5
#train the model
knnmodel.fit(x_train,y_train)# here model train, then after training the model it is ready for testing
```

```python
y_pred =knnmodel.predict(x_test) #knnmoedl is the nickname for the Kneighbot=rs Regressor , prefdict for testigng , after moddel fitting
```

```python
import math # math librarry to import # for regressor , mean error, which is represented as the +-error  mean error
from sklearn.metrics import mean_squared_error
meanerro = math.sqrt(mean_squared_error(y_pred,y_test))
print(f"the value of mean squared error is {meanerro}")
```

```python
#evaluation for increase accuracy  or decrese error
from sklearn.metrics import root_mean_squared_error
error =root_mean_squared_error(y_pred,y_test)
```
